```python
from transformers import pipeline
from datasets import load_dataset
from tqdm import tqdm
```

```python
ds = load_dataset('csv', data_files='/content/drive/MyDrive/data/de_en_filtered.csv')
ds = ds['train']
ds
```

```python
pipe = pipeline('translation', model='Helsinki-NLP/opus-mt-de-en', device=0)
```

```python
def data():
  for row in ds:
    yield row['German']
```

```python
ds_length = ds.num_rows
ps = []
for res in tqdm(pipe(data(), batch_size=48), total=ds_length):
  #print(res)
  ps.append(res[0]['translation_text'])

```

```python
ds = ds.add_column(name='Paraphrase', column=ps)
ds = ds.rename_column('English', 'Original')
ds = ds.remove_columns(['German'])
ds = ds.to_csv('/content/drive/MyDrive/data/paraphrases_en.csv')
```
