## Data Download and Preparation

```python
from datasets import load_dataset
```

We use our own created datasets with paraphrases

```python
raw_ds_enen = load_dataset(enen, split='test')
raw_ds_slsl = load_dataset(slsl, split='test')
raw_ds_cscs = load_dataset(cscs, split='test')
raw_ds_dede = load_dataset(dede, split='test')
```

We load the evaluated multilanguage dataset

```python
raw_ds_multi = load_dataset("csv", data_files=dataset, split='train')
raw_ds_multi, raw_ds_enen, raw_ds_slsl, raw_ds_cscs, raw_ds_dede
```

## Splitting the multilingual dataset

Initialize the new datasets

```python
import tensorflow as tf
device_name = tf.test.gpu_device_name()
if len(device_name) > 0:
    print("Found GPU at: {}".format(device_name))
else:
    device_name = "/device:CPU:0"
    print("No GPU, using {}.".format(device_name))
```

```python
with tf.device(device_name):
  ds_multi_enen = raw_ds_multi.filter(lambda row: row['Original'] in raw_ds_enen['Original'])
```

```python
with tf.device(device_name):
  ds_multi_slsl = raw_ds_multi.filter(lambda row: row['Original'] in raw_ds_slsl['Original'])
```

```python
with tf.device(device_name):
  ds_multi_cscs = raw_ds_multi.filter(lambda row: row['Original'] in raw_ds_cscs['Original'])
```

```python
with tf.device(device_name):
  ds_multi_dede = raw_ds_multi.filter(lambda row: row['Original'] in raw_ds_dede['Original'])
```

## Calculate average parascore and save the datasets

Export the tables

```python
ds_multi_enen.to_csv('/content/drive/MyDrive/data/eval_table_multi_small_enen.csv')
ds_multi_slsl.to_csv('/content/drive/MyDrive/data/eval_table_multi_small_slsl.csv')
ds_multi_cscs.to_csv('/content/drive/MyDrive/data/eval_table_multi_small_cscs.csv')
ds_multi_dede.to_csv('/content/drive/MyDrive/data/eval_table_multi_small_dede.csv')
```

Calculate the average scores

```python
score_enen = sum(ds_multi_enen['Parascore']) / len(ds_multi_enen)
print('Average Parascore EnEn:', score_enen)
```

```python
score_slsl = sum(ds_multi_slsl['Parascore']) / len(ds_multi_slsl)
print('Average Parascore SlSl:', score_slsl)
```

```python
score_cscs = sum(ds_multi_cscs['Parascore']) / len(ds_multi_cscs)
print('Average Parascore CsCs:', score_cscs)
```

```python
score_dede = sum(ds_multi_dede['Parascore']) / len(ds_multi_dede)
print('Average Parascore DeDe:', score_dede)
```
